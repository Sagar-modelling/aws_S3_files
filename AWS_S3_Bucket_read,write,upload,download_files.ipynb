```python
import pandas as pd

# Make dataframes in S3 Bucket to write a file
df1 = pd.DataFrame({'x': [1, 2, 3], 'y': ['a', 'b', 'c']})
df2 = pd.DataFrame({'x': [10, 20, 30], 'y': ['aa', 'bb', 'cc']})

# Save to csv
df1.to_csv('df1.csv')
df2.to_csv('df2.csv')
```

```python
# Upload files to S3 bucket
s3.Bucket('dataframe1').upload_file(Filename='df1.csv', Key='df1.csv')
s3.Bucket('dataframe1').upload_file(Filename='df2.csv', Key='df2.csv')
```

```python
# Load csv file directly into python
obj = s3.Bucket('dataframe1').Object('df1.csv').get()
deep = pd.read_csv(obj['Body'], index_col=0)
```

```python
# Download file and read from disc
s3.Bucket('dataframe1').download_file(Key='df1.csv', Filename='download_bucket_file')
pd.read_csv('download_bucket_file', index_col=0)
```
